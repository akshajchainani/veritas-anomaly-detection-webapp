```python
import pandas as pd
features = pd.read_csv('/kaggle/input/nsw-full/NB15_features.csv',engine='python')
df_1 = pd.read_csv('/kaggle/input/nsw-full/UNSW-NB15_1.csv')
col_names = features['Name']
df_1.columns = col_names
df_2 = pd.read_csv('/kaggle/input/nsw-full/UNSW-NB15_2.csv')
df_2.columns = col_names
df_3 = pd.read_csv('/kaggle/input/nsw-full/UNSW-NB15_3.csv')
df_3.columns = col_names
df_4 = pd.read_csv('/kaggle/input/nsw-full/UNSW-NB15_4.csv')
df_4.columns = col_names
df = pd.concat([df_1,df_2,df_3,df_4],axis=0)
print(len(df.columns))
print(df.columns)
```

```python
X = df.drop(columns='Label')
X = df[['proto','Spkts','Dpkts','tcprtt','state','dur','sbytes','dbytes','ct_srv_src','ct_srv_dst']] #'sport','dsport'
```

```python
y = df['Label']
y.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
clf = RandomForestClassifier(class_weight='balanced', verbose=1)
```

```python
X = pd.get_dummies(X,columns=['proto','state'])
```

```python
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, make_scorer
import pickle
with open('/kaggle/input/models/randomforests_final2.pkl','rb') as f:
    clf = pickle.load(f)
scores = cross_validate(clf,X,y,cv=5,scoring=['recall_macro','precision_macro','f1_macro','accuracy'])
print(scores.keys())
```

```python
preds = clf.predict(X)
print(accuracy_score(preds,y))
print(f1_score(preds,y))
print(recall_score(preds,y))
print(precision_score(preds,y))
```

```python
print(scores['test_recall_macro'])
print(scores['test_accuracy'])
print(scores['test_f1_macro'])
print(scores['test_precision_macro'])
```
